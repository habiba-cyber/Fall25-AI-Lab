```python
film_list = [
    ("Eternal Sunshine of the Spotless Mind", 20000000),
    ("Memento", 9000000),
    ("Requiem for a Dream", 4500000),
    ("Pirates of the Caribbean: On Stranger Tides", 379000000),
    ("Avengers: Age of Ultron", 365000000),
    ("Avengers: Endgame", 356000000),
    ("Incredibles 2", 200000000)
]

extra_count = int(input("How many films do you want to add? "))

for _ in range(extra_count):
    film_title = input("Enter film title: ").strip()
    film_cost = int(input("Enter film budget: "))
    film_list.append((film_title, film_cost))

# average cost
total_cost = sum(cost for _, cost in film_list)
avg_cost = total_cost / len(film_list)
print(f"\nAverage Budget: ${avg_cost:,.0f}")

# films above average
above_count = 0
print("\nFilms with budget above average:\n")
for title, cost in film_list:
    if cost > avg_cost:
        diff = cost - avg_cost
        print(f"{title} - ${diff:,.0f} above average")
        above_count += 1

print(f"\nNumber of films above average: {above_count}")
```
