```python
import random

def game():
   
   words = ["python", "computer"]
   hide_word = random.choice(words)  # random word choose
   display = ["_"] * len(hide_word)       # hidden letters
   trials = 6                              # max wrong attempts

   lost_cases = [
        """
           ----
           |  |
           |  O
           | /|\\
           | / \\
           |
        """,
        """
           ----
           |  |
           |  O
           | /|\\
           | /
           |
        """,
        """
           ----
           |  |
           |  O
           | /|\\
           |
           |
        """,
        """
           ----
           |  |
           |  O
           | /|
           |
           |
        """,
        """
           ----
           |  |
           |  O
           |
           |
           |
        """,
        """
           ----
           |  |
           |
           |
           |
           |
        """,
        """
           ----
           |
           |
           |
           |
           |
        """,
    ]

   print("Welcome to Hangman")
   print("Time to play!!")

   while trials > 0 and "_" in display:
      print("\nWord: ", " ".join(display))
      print(lost_cases[6 - trials])
      letter = input("Guess a letter: ").lower()

      if letter in hide_word:
            print("Wow! Correct guess")
            for i, ch in enumerate(hide_word):
                if ch == letter:
                    display[i] = letter
      else:
            trials -= 1
            print("Oops! You lost a chance, Chances left:", trials)

   if "_" not in display:
        print("\nCongratulations, the shot was on the point! The word was:", hide_word)
   else:
        print(lost_cases[0])
        print("\nGame Over! The word was:", hide_word)

game()
```
