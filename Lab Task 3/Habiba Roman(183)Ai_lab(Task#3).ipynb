```python
class BasicHeaterAgent:
    def __init__(self, setpoint=22.0):
        self.setpoint = float(setpoint)
        self.current_reading = None

    def update_reading(self, value):
        self.current_reading = float(value)

    def act_decide(self):
        return "HEATER_ON" if self.current_reading < self.setpoint else "HEATER_OFF"

    def execute_action(self, room):
        action = self.act_decide()
        print(f"[Basic] {room} | Temp={self.current_reading}°C -> {action}")
        return action


class SmartHeaterAgent:
    def __init__(self, setpoint=22.0, margin=0.5):
        self.setpoint = float(setpoint)
        self.margin = float(margin)
        self.current_reading = None
        self.previous_state = "HEATER_OFF"

    def update_reading(self, value):
        self.current_reading = float(value)

    def act_decide(self):
        lower_bound = self.setpoint - self.margin
        upper_bound = self.setpoint + self.margin
        if self.current_reading < lower_bound:
            return "HEATER_ON"
        if self.current_reading >= upper_bound:
            return "HEATER_OFF"
        return self.previous_state

    def execute_action(self, room):
        action = self.act_decide()
        if action != self.previous_state:
            print(f"[Smart] {room} | Temp={self.current_reading}°C -> {action} (changed)")
        else:
            print(f"[Smart] {room} | Temp={self.current_reading}°C -> {action} (no change)")
        self.previous_state = action
        return action


if __name__ == "__main__":
    rooms_temp = {
        "Living Room": [20.0, 21.0, 21.7, 22.1, 21.9],
        "Kitchen": [24.0, 23.5, 23.0, 22.4, 22.0],
        "Bedroom": [18.5, 19.8, 20.5, 21.6, 22.2],
        "Study Room": [19.0, 20.0, 20.8, 21.5, 22.3],
        "Dining Room": [21.0, 21.5, 22.0, 22.6, 22.1],
    }

    print("Basic Heater Agent")
    basic_agent = BasicHeaterAgent(setpoint=22.0)
    for room, readings in rooms_temp.items():
        print(f"\nRoom: {room}")
        for temp in readings:
            basic_agent.update_reading(temp)
            basic_agent.execute_action(room)

    print("\nSmart Heater Agent")
    smart_agent = SmartHeaterAgent(setpoint=22.0, margin=0.5)
    for room, readings in rooms_temp.items():
        print(f"\nRoom: {room}")
        smart_agent.previous_state = "HEATER_OFF"
        for temp in readings:
            smart_agent.update_reading(temp)
            smart_agent.execute_action(room)
```
